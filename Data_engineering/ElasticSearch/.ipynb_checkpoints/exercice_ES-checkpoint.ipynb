# Importer les donn√©es

```python
from elasticsearch import Elasticsearch

LOCAL = False

es_client = Elasticsearch(hosts=["localhost" if LOCAL else "elasticsearch"])
```

```python
import pandas as pd 
import numpy as np
from elasticsearch.helpers import bulk
```

```python
import pandas as pd
import json
df_movies = pd.read_csv("./data/tmdb_5000_movies.csv")
```

```python
df_credits = pd.read_csv("data/tmdb_5000_credits.csv")
```

```python
df_credits.head()
```

```python
df_movies.head()
```

## 1) Nettoyer les donn√©es de la DataFrame d'acteurs comme pr√©c√©dement.  

```python
def clean_words(l):
    return [elt["name"] for elt in json.loads(l)]
```

```python
for col in ["genres", "keywords", "production_countries", "spoken_languages"]:
    df_movies.loc[:,col] = df_movies.loc[:,col].apply(clean_words)
```

```python
for col in ["cast", "crew"]:
    df_credits.loc[:,col] = df_credits.loc[:,col].apply(clean_words)
```

```python
df_credits.rename(columns={"movie_id":"id"},inplace=True)
```

## 2) Merger les deux DataFrame en utilisant l'identifiant du film.  

```python
df=df_movies.join(df_credits.set_index('id'), on='id',how='left', lsuffix='_movies', rsuffix='_credits')
```

```python
df.drop(columns=["title_credits"],inplace=True)
```

```python
df.head()
```

## 3) Cr√©er un nouvel Index `augmented_movies` similaire √† l'index pr√©c√©dant mais en ajoutant les donn√©es des acteurs.

```python
documents1 = df.fillna("").to_dict(orient="records")
```

```python
documents1[0:2]
```

```python
def generate_data(documents):
    for docu in documents:
        yield {
            "_index": "augmented_movies",
            "_type": "movie",
            "_source": {k:v if v else None for k,v in docu.items()},
        }

a=bulk(es_client, generate_data(documents1))
```

```python
es_client.ping()
```

```python
result = es_client.search(index="augmented_movies", body={"query": {"match_all": {}}})
result
```

## 4) Cr√©er une fonction permettant de trouver tous les films d'un acteur. 

```python
QUERY = {
    "_source":"title_movies",
    "query": {
        "match":{"cast":"Depp"}
        
    }
}
```

```python
result = es_client.search(index="augmented_movies", body=QUERY)
result
```

```python
def fun(x):
    QUERY = {
    "_source":"title_movies",
    "query": {
        "match":{"cast":x}
        
    }
}
    return QUERY
```

```python
result = es_client.search(index="augmented_movies", body=fun("Depp"))
result
```

## 5) Trouver quel acteur √† jouer dans les films avec les plus gros budgets.

```python
QUERY = {
    "_source":["title_movies","cast","budget"],
    "size":3,
    "sort" : [
        { "budget" : {"order" : "desc"}}
    ]
}
```

```python
result = es_client.search(index="augmented_movies", body=QUERY)
result
```
