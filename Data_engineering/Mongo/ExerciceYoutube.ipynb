```python
from datetime import datetime
df_youtube = pd.read_csv("./data/USvideos.csv",parse_dates=['date'])
df_youtube['date']=df_youtube['date']+'.2020'
df_youtube['date']=df_youtube['date'].apply(lambda x:datetime.strptime(str(x),"%d.%M.%Y"))
df_youtube.head()
```

## Question 0

```python
import json
with open("./data/US_category_id.json") as readfile:
    data = json.load(readfile)
    
data= pd.json_normalize(data['items'])
data.head()
```

```python
data['id']=data['id'].astype(int)
```

```python
df=df_youtube.join(data.set_index('id'), on='category_id')
df.head()
```

### Netoyer les données

```python
df.isna().sum().sum()
```

```python
df.columns
```

```python
df=df.rename(columns={"snippet.title":"snippet_title",'snippet.assignable':'snippet_assignable','snippet.channelId':'snippet_channelId'})
```

```python
df['ID'] = df.index
```

```python
df.head()
```

### Importer les données

```python
def slice_generator(df, chunk_size=10):
    current_row = 0
    total_rows = df.shape[0]
    while current_row < total_rows:
        yield df[current_row:current_row + chunk_size]
        current_row += chunk_size

    
```

```python
for df_chunk in slice_generator(df):
    records = df_chunk.to_dict(orient='records')
    collection.insert_many(records)
```

## Question 3

```python
map_function = '''function(){emit(this.video_id, this.tags)}'''
reduce_function = '''function(key,values){return values.toString().split("|")}'''
out = '''{ out:"tags"}'''
result = collection.map_reduce(
        map_function, 
        reduce_function,
        out)
```

```python
dictio=list(result.find())
```

```python
dictio[1]['value']
```

```python
len(dictio)
```

```python
for i in range(len(dictio)):
    collection.update_one({'video_id':dictio[i]["_id"]},{"$set":{'tags':dictio[i]['value']}})
    
    
    
```

## Question 4

```python
cur = collection.aggregate([{"$group" : {"_id" : "$video_id", "averageviewsnumber" : {"$avg" : "$views"}}},{"$sort":{"averageviewsnumber":-1}},{"$limit":5}])
print("les vidéos les plus vues sont :")
list(cur)
```
